```python
## predictive inference
with open("ARM_data/child.iq/kidiq.dta") as f:
    df = pd.read_stata(f)
msk = np.hstack((True,np.zeros(len(df)-1,'bool')))
np.random.seed(0)
np.random.shuffle(msk)
df[msk]
```

```python
lm = smf.ols('kid_score ~ mom_hs+mom_iq+C(mom_work)+mom_age', data=df[~msk]).fit()
lm.predict(df[msk])
```

```python
pred = lm.get_prediction(df[msk])
print 'prediction = ',pred.predicted_mean
print 'confidence interval for E[Y|X] = ',pred.conf_int(alpha=0.05)
print 'confidence interval for Y      = ',pred.conf_int(alpha=0.05,obs=True)
```

```python
np.random.seed(0)
hits = 0
width = 0
msk = np.hstack((np.zeros(50,'bool'),np.ones(len(df)-50,'bool')))
for i in range(100):
    np.random.shuffle(msk)
    lm = smf.ols('kid_score ~ mom_hs+mom_iq+C(mom_work)+mom_age', data=df[~msk]).fit()
    ci = lm.get_prediction(df[msk]).conf_int(alpha=0.05,obs=True)
    width += (ci[:,1]-ci[:,0]).mean()
    hits += ((df[msk]['kid_score'] >= ci[:,0]) & (df[msk]['kid_score'] <= ci[:,1])).mean()
print hits / 100.
print width / 100.
```

```python
with open("diabetes.tab") as f:
    diabetes = pd.read_table(f)
diabetes.head()
```

```python
singlefeatures=list(set(diabetes.columns)-{'y'})
print singlefeatures
```

```python
import itertools
interactions = ['%s:%s'%v for v in itertools.combinations(singlefeatures,2)]
print interactions[:10]
```

```python
lm = smf.ols('y ~ '+'+'.join(singlefeatures), data=diabetes).fit()
lm.summary()
```

```python
lm.aic
```

```python
def fitmodel(S,data):
    return smf.ols('y ~ '+('+'.join(S) if len(S)>0 else '1'), data=data).fit()
```

```python
## forward stepwise
Sfwd = set()
# features = set(singlefeatures)
features = set(singlefeatures).union(set(interactions))
while len(Sfwd)<len(features):
    f = max(features - Sfwd, key = lambda f: fitmodel(Sfwd.union({f}),diabetes).aic)
    after = fitmodel(Sfwd.union({f}),diabetes).aic
    before = fitmodel(Sfwd,diabetes).aic
    if after > before:
        Sfwd = Sfwd.union({f})
    else:
        break
print Sfwd
    
```

```python
## backward stepwise
# features = set(singlefeatures)
features = set(singlefeatures).union(set(interactions))
Sbwd = set(features)
while len(Sbwd)>0:
    f = max(Sbwd, key = lambda f: fitmodel(Sbwd-{f},diabetes).aic)
    after = fitmodel(Sbwd-{f},diabetes).aic
    before = fitmodel(Sbwd,diabetes).aic
    if after > before:
        Sbwd = Sbwd-{f}
    else:
        break
print Sbwd
    
```

```python
## stepwise
Sboth = Sfwd if fitmodel(Sfwd,diabetes).aic > fitmodel(Sbwd,diabetes).aic else Sbwd
print Sboth
```
