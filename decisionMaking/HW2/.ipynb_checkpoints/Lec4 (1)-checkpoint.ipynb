```python
%pylab
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
sns.set()
import pandas as pd
import scipy as sp
import scipy.stats
```

```python
def generate_data(shape):
    return sp.stats.norm.rvs(size=shape)
```

```python
## student t
np.random.seed(2)
n = 10
dataset = generate_data(n)
print 'true mean = 0'
print 'sample mean =',dataset.mean()
```

```python
mu0 = 1
tn  = (dataset.mean()-mu0)*np.sqrt(n)/dataset.std(ddof=1)
p   = 2*(1-scipy.stats.t.cdf(abs(tn),n-1))
print 'p-val for mu0=%f is %f'%(mu0,p)
```

```python
## t confidence intervals
Q = sp.stats.t.ppf(1-.05/2,n-1)
print 'confidence interval: [%f, %f]'%(dataset.mean()-Q*dataset.std(ddof=1)/np.sqrt(n),dataset.mean()+Q*dataset.std(ddof=1)/np.sqrt(n))
```

```python
## using statsmodels
import statsmodels.stats.api as sms
print 'tn,p,dof =',sms.DescrStatsW(dataset).ttest_mean(mu0)
print 'CI =',sms.DescrStatsW(dataset).tconfint_mean()
```

```python
with open("ARM_data/child.iq/kidiq.dta") as f:
    df = pd.read_stata(f)
```

```python
print df.kid_score.mean()
```

```python
zip(df.columns,np.array(sms.DescrStatsW(df).tconfint_mean()).T.tolist())
```

```python
## KS
q = scipy.stats.ksone.ppf(1.-.05/2., len(df.kid_score))
plot(np.sort(df["kid_score"]), np.arange(len(df["kid_score"]))/float(len(df["kid_score"])))
plot(np.sort(df["kid_score"]), np.maximum(0,np.arange(len(df["kid_score"]))/float(len(df["kid_score"]))-q),'c')
plot(np.sort(df["kid_score"]), np.minimum(1,np.arange(len(df["kid_score"]))/float(len(df["kid_score"]))+q),'c')
```

```python
## test normality
print sms.diagnostic.kstest_normal(df.kid_score)
```

```python
## qq plot
import statsmodels.api as sm
fig = sm.graphics.qqplot(df.kid_score, sp.stats.norm, fit=True, line='45')
```
